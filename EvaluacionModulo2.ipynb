## Descripción del proyecto

Este proyecto consiste en el desarrollo de un sistema de gestión de contactos en Python. 
La aplicación permite registrar, buscar, editar y eliminar contactos utilizando 
Programación Orientada a Objetos y estructuras de datos como listas y diccionarios, 
con el objetivo de organizar la información de manera clara y eficiente.

## Guía de uso

1. Crear un objeto de la clase GestorContactos.
2. Crear contactos usando la clase Contacto.
3. Agregar contactos al gestor.
4. Utilizar los métodos para buscar, editar o eliminar contactos.

```python
class Contacto:
    def __init__(self, nombre, telefono, correo, direccion):
        self.nombre = nombre
        self.telefono = telefono
        self.correo = correo
        self.direccion = direccion

    def to_dict(self):
        return {
            "nombre": self.nombre,
            "telefono": self.telefono,
            "correo": self.correo,
            "direccion": self.direccion
        }
```

```python
class GestorContactos:
    def __init__(self):
        self.contactos = []

    def agregar_contacto(self, contacto):
        self.contactos.append(contacto.to_dict())

    def listar_contactos(self):
        return self.contactos
    
    def buscar_contacto(self, dato):
        resultados = []
        for c in self.contactos:
            if dato.lower() in c["nombre"].lower() or dato in c["telefono"]:
                resultados.append(c)
        return resultados
    
    def editar_contacto(self, telefono, nuevos_datos):
        for c in self.contactos:
            if c["telefono"] == telefono:
                c.update(nuevos_datos)
                return True
        return False
    
    def eliminar_contacto(self, telefono):
        for c in self.contactos:
            if c["telefono"] == telefono:
                self.contactos.remove(c)
                return True
        return False
```

```python
gestor = GestorContactos()

c1 = Contacto("Ana", "1234", "ana@mail.com", "Calle 1")
c2 = Contacto("Luis", "5678", "luis@mail.com", "Calle 2")

gestor.agregar_contacto(c1)
gestor.agregar_contacto(c2)

print("Listado:", gestor.listar_contactos())
print("Buscar Ana:", gestor.buscar_contacto("Ana"))

gestor.editar_contacto("1234", {"correo": "nuevo@mail.com"})
gestor.eliminar_contacto("5678")

print("Final:", gestor.listar_contactos())
```
